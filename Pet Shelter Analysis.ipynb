# 2.What is the average number of pets found in a month in the year 2015? Are there months where there is a higher number of animals found?

Knowing the number of pets the shelter might see in a month can help them gather enough resources and donations to care for the animals they receive.


## The average number of pets found in the year 2015 is 1559.33.The highest number of pets are found in May and June.


## Explanation

Filtered the intakes table to create a new dataframe 'avg_num_pets' to collect data between the year 2015 and 2016.Created a new column to display the month. Then grouped the data by month using the groupby()function. After that, I applied  the size () function to show how many pets found per month.
Finally, the average was calculated using mean() to the dataframe results.

```python
intakes.head(1)
```

```python
avg_num_pets=intakes.loc[(intakes.datetime >='2015')& (intakes.datetime < '2016')]

avg_num_pets.head(5)
```

```python
avg_num_pets=intakes.loc[(intakes.datetime >='2015')& (intakes.datetime < '2016')]

avg_num_pets['Month']= avg_num_pets['datetime'].str[5:7]

avg_num_pets['Month']=avg_num_pets['Month'].astype('int32')

avg_num_pets.head()
```

```python
groupby=avg_num_pets.groupby('Month')

results=groupby.size()

results

```

```python
results.mean()
```

## 7. About how many animals are spayed/neutered each month?

This will help the shelter allocate resources and staff.Assume that all intact males and females will be spayed /neutered.

898 animals are spayed and neutered each month between the year 2017 and 2018 except for the months May and June where there is a spike in the amont of intakes


# Explanation

Used data from the last year 2017 - 2018.
Filtered the relevant columns and animals that are intact. Ignored NAN data.
Grouped the dataframe 'spayedNeutered' by month to create a new dataframe.Used size() function to calculate the number of intact animals in that year.
Finally applied mean() function to calculate the average.

```python
#Calculating the most recent year
intakes_outcomes= intakes_outcomes.sort_values('intake_year', ascending=False)
intakes_outcomes.head(5)
```

```python
spayedNeutered=intakes_outcomes.loc[(intakes_outcomes.intake_datetime >='2017')& (intakes_outcomes.intake_datetime < '2018')]
spayedNeutered = spayedNeutered.loc[:,['animal_type','sex_upon_intake','intake_month','intake_year']]
spayedNeutered=spayedNeutered.loc[spayedNeutered['sex_upon_intake'].str.contains('Intact',na=False)] 

spayedNeutered
```

#### Ignored   NA/NaN value to fix error " ValueError: Cannot mask with non-boolean array containing NA / NaN values"

```python
spayedNeutered.count() # No missing data 
```

```python
bymonth=spayedNeutered.groupby('intake_month')
bymonth=bymonth.size()
bymonth
```

```python
bymonth.mean()
```
